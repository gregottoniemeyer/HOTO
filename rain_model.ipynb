```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Number of rows
num_rows = 7200

# Initialize rain column with all zeros
rain = np.zeros(num_rows)

# Define the start, peak, and end points of the rain wave
start_rain = 1200
peak_rain = 1250
end_rain = 1350

# Generate easing in and out using a sine function for smooth transition
wave_length = end_rain - start_rain
x = np.linspace(0, np.pi, wave_length)
rain_wave = np.sin(x) ** 2  # Eases in and out smoothly

# Set the rain values in the defined range
rain[start_rain:end_rain] = rain_wave

# Create DataFrame
rain_data = pd.DataFrame({'rain': rain})
```

```python
# Plot the rain column to visualize
plt.plot(rain_data['rain'][start_rain:end_rain])
plt.xlabel('Row Index')
plt.ylabel('Rain Intensity')
plt.title('Rain Wave with Ease In and Ease Out')
plt.show()
```

```python
# Save to CSV (optional)
rain_data.to_csv('./data/rain_01.csv', index=False)
```
