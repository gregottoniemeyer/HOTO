```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#get csv files
#out: merged_df.to_csv("../data/4min_pass_wave_salt_temp.csv", index=True)
data_location_01 = "../data/data_bottom_pause_Salt_nolatlon.csv"
df_salinity = pd.read_csv(data_location_01)
data_location_02 = "../data/sine_02.csv"
df_wave = pd.read_csv(data_location_02)
data_location_03 = "../data/dive.csv"
df_temp = pd.read_csv(data_location_03)

#new for 12 min version
data_location_04 = "../data/creatures.csv"
df_creatures = pd.read_csv(data_location_04)

data_location_05 = "../data/buoy_temp_height.csv" #buoy Wave Height
df_buoy = pd.read_csv(data_location_05)

data_location_06 = "../data/global_temp.csv" #buoy surface temo
df_global = pd.read_csv(data_location_06)

data_location_07 = "../data/clouds.csv" #buoy surface temo
df_clouds = pd.read_csv(data_location_07)
```

```python
col1 = df_temp['Depth']  # Column from df1
col2 = df_wave['sine_wave']  # Column from df1
col3 = df_salinity['Salinity']  # Column from df2
col4 = df_temp['Temp']  # Column from df2
col5 = df_creatures['surface creatures start']  # Column from df1
col6 = df_creatures['wind mix creatures']  # Column from df1
col7 = df_creatures['midnight creatures']  # Column from df2
col8 = df_creatures['surface creatures end']  # Column from df2

merged_df = pd.DataFrame({'depth': col1, 'wave': col2, 'salinity': col3, 'temp': col4, 'surface life down': col5, 'windmix life': col6, 'midnight life': col7, 'surface life up': col8})
normalized_df = merged_df.apply(lambda x: (x - x.min()) / (x.max() - x.min()))

normalized_df["buoy_height"] = df_buoy["wave_height"]
normalized_df["buoy_temp"] = df_buoy["air_temp"]
normalized_df["global_temp"] = df_global["sst"]
normalized_df.fillna(0.0, inplace=True)
```

```python
normalized_df.tail()
```

```python
fig, ax1 = plt.subplots(figsize=(16, 4))
ax1.plot(normalized_df.index, normalized_df['buoy_height'], color="dodgerblue")
ax1.set_ylabel('Wave Height', color="dodgerblue")
ax1.set_xlabel('Index')
ax2 = ax1.twinx()
ax2.plot(normalized_df.index, normalized_df['buoy_temp'], color="orange")
ax2.set_ylabel('Air Temperature', color="orange")
plt.savefig("../img/buoy.png")
plt.show()
```

```python
normalized_df[df_clouds.columns] = df_clouds
normalized_df.head()
```

```python
normalized_df = normalized_df.drop(columns=['Unnamed: 0'])
normalized_df.head()
```
