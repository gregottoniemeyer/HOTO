```python
from goes2go import GOES
import warnings
import numpy as np
warnings.filterwarnings('ignore')

#list of GOES products https://github.com/blaylockbk/goes2go/blob/main/goes2go/product_table.txt
```

```python
# ABI Multi-Channel Cloud Moisture Imagry Product
G = GOES(satellite=16, product="ABI-L2-RRQPEF", domain='F')
# Download and read the data as an xarray Dataset nearest a specific time
ds = G.nearesttime('2024-06-22')
ds
```

```python
tem = np.log(ds.RRQPE)
tem.plot(vmin=0,vmax=2,cmap='turbo')
```

```python
# ABI Multi-Channel Cloud Moisture Imagry Product
G = GOES(satellite=16, product="ABI-L2-SSTF", domain='F')
# Download and read the data as an xarray Dataset nearest a specific time
ds = G.nearesttime('2024-06-22')
ds
```

```python
tem = ds.SST-273.15
tem.plot(vmin=-3,vmax=30,cmap='turbo')
```

```python
tem.isel(y=slice(1000,2000),x=slice(2000,4000)).plot(vmin=-3,vmax=30,cmap='turbo')
```
